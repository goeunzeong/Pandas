**4. 아래 함수를 적용하여, 업종을 라벨링 하시오.**

```python
def div_func(data):
    if data==70:
        return '의료기관'
    elif data == 71 :
        return '보건위생'
    elif data== 21 :
        return '레저업소'
    elif data == 22:
        return '문화취미'
    elif data== 50 :
        return '서적문구'
    elif data ==42:
        return '의복'
    elif data ==33:
        return '연료판매'
    elif data == 62:
        return '자동차정비'
    elif data == 92:
        return '수리서비스'
    elif data == 44:
        return '잡화'
    elif data==40:
        return '유통업'
    elif data ==80:
        return '요식업소'
    elif data ==81:
        return '음료식품'
    else:
        return '기타'
```

```python
df1['업종명'] = df1['업종'].apply(div_func)
df1
```

**5. 구매금액의 단위를 맞춰 변환하시오. (현재 파일 단위 : 천원)**

```python
df1['이용금액(원)'] = df1['이용금액'] * 1000
df1
```

**6. 업종별로 구매금액의 총합과 빈도수(count)를 계산하여 엑셀파일형태로 저장하시오. (파일제출시 같이 제출)**

```python
df1['빈도수'] = 1
df1_pv = df1.pivot_table(index = '업종명', values = ['이용금액(원)', '빈도수'], aggfunc='sum').reset_index()
df1_pv.to_excel('5차_업종별_구매금액총합_빈도수.xlsx')
df1_pv

# aggfunc = ['sum', 'count'] : 라고 해도 빈도수와 합 모두 계산 가능.
```

**7. 기준일자를 datetime 형태로 변환하시오**

```python
df1['기준일자(dt)'] = pd.to_datetime(df1['기준일자'], format = '%Y%m%d')
df1
```

**8. 변환된 기준일자를 이용하여, 연도, 월 , 요일, 주차, 일자를 계산해, 새로운 Column으로 선언하시오**

```python
df1['기준연도'] = df1['기준일자(dt)'].dt.year
df1['기준월'] = df1['기준일자(dt)'].dt.month
df1['기준일'] = df1['기준일자(dt)'].dt.day
df1['기준요일'] = df1['기준일자(dt)'].dt.day_name()
df1['기준주차'] = df1['기준일자(dt)'].dt.isocalendar().week
df1
```

**9. 각 월별, 요일별 이용금액의 합을 계산하여, 엑셀파일형태로 저장하시오. (저장된 파일을 같이 제출)**

```python
df1_pv9 = df1.pivot_table(index = '기준월', values = '이용금액(원)', columns = '기준요일', aggfunc='sum')
df1_pv9.to_excel('5차_월별_요일별_이용금액의_합.xlsx')
df1_pv9
```

**10. 업종 중, '의복'의 Column만 추출하여, 성별에 따른 이용금액의 합을 계산해, 엑셀파일 형태로 저장하시오. (저장된 파일을 같이 제출)**

```python
df1_clothes = df1.loc[(df1['업종명'] == '의복')]
df1_clothes_pv = df1_clothes.pivot_table(index = '성별', values = '이용금액(원)', aggfunc = 'sum').reset_index()
df1_clothes_pv.to_excel('성별에_따른_의복_이용금액의_총합.xlsx')
```

**11. '주소-구' Column의 값을 아래를 참조하여 변환하시오.**

* 종로구 : 110 
* 노원구 : 350 

```python
df1['주소-구'].value_counts()
```

```python
cond1 = ( df1['주소-구'] == 110 )
df1.loc[cond1, '주소-구명'] = '종로구'
df1.loc[~cond1, '주소-구명'] = '노원구'
df1
```

**12. 이용금액에 이용건수를 나눠, 건당 이용 금액을 계산하시오**

```python
df1['건당이용금액'] = df1['이용금액(원)'] / df1['이용건수']
df1
```

**13. 종로구의 업종별 건당이용금액의 합을 계산하여, 가장 높은 건당 이용금액의 업종을 파악하시오.**

```python
df1_jr = df1.loc[(df1['주소-구명'] == '종로구')]
df1_jr.pivot_table(index = '업종명', values = '건당이용금액', aggfunc = 'sum').reset_index().sort_values(by ='건당이용금액', 
                                                                                                ascending = False)
# 1위: 기타, 2위: 의료기관
```

**14. 노원구의 업종별 건당이용금액의 합을 계산하여, 가장 높은 건당 이용금액의 업종을 파악하시오.**

```python
df1_nw = df1.loc[(df1['주소-구명'] == '노원구')]
df1_nw.pivot_table(index = '업종명', values = '건당이용금액', aggfunc = 'sum').reset_index().sort_values(by ='건당이용금액', 
                                                                                                ascending = False)
# 1위: 기타, 2위: 보건위생
```

**15. 성별과 주소(노원/종로), 나이에 따른 이용금액의 합을 계산하여, 엑셀파일 형태로 저장하시오 (저장된 파일 같이 제출)**

```python
df1_pv15 = df1.pivot_table(index = ['주소-구명', '성별'], values = '이용금액(원)', columns = '나이', aggfunc='sum')
df1_pv15.to_excel('5차_성별_주소_나이에_따른_이용금액의_합.xlsx')
df1_pv15
```
