**5.회원가입일을 기준으로 회원가입 연도 /월 /주차 / 요일 을 계산하시오.**

```python
df1['회원가입일(dt)'] = pd.to_datetime(df1['회원 가입일'])
df1['가입연도'] = df1['회원가입일(dt)'].dt.year
df1['가입월'] = df1['회원가입일(dt)'].dt.month
df1['가입주차'] = df1['회원가입일(dt)'].dt.isocalendar().week
df1['가입요일명'] = df1['회원가입일(dt)'].dt.day_name()
df1['가입요일'] = df1['회원가입일(dt)'].dt.weekday
df1
```

**8. 분류모델을 적용하기위해, Y(Target) Table과 X Table을 새로 선언하시오.**

* Y = 평생회원여부
* X = 회원가입월 /회원가입연도 /회원가입요일 /나이 / 누적주문건수 /실결제금액 /총 사용 적립금 

```python
Y = df1_dn['평생회원']
X = df1_dn[['가입월', '가입연도', '가입요일', '나이', '누적주문건수', '실결제금액', '총 사용 적립금']]
```

**10.Sklearn 라이브러리 활용, 위에서 정의한 X,Y Table에 대해 Train Set과 Test Set으로 분할하여 4개의 Table의 Column수와 Row수를 확인하시오.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.2)
```

**11.Sklearn 라이브러리의 Decision Tree Classification Model을 불러와 Train Set에 대해 Modeling을 실시하시오.**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

**12.Model을 바탕으로 각각 X Train Set과 X Test Set에 대해, 예측 Y 값을 계산하시오.**

```python
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)
```

**13. Model에 의해 예측된 값을 바탕으로 Train과 Test의 Model 정확도를 계산하시오.**

(Sklearn을 활용해 계산)

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_train, Y_train_pred)
```

```python
accuracy_score(Y_test, Y_test_pred)
```
